```python
import numpy as np
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
#Importing Dataset
df = pd.read_csv('time_series_covid19_confirmed_global.csv')
df_india = df.loc[df["Country/Region"] == 'India']
df1_transposed = df_india.T
df_india = df1_transposed.loc['1/22/20':]
df_india = df_india.rename(columns={131:"India"})
df_india
```

```python
# Spitting into train and test data
x_train = df_india
y_train = df_india

model = GradientBoostingRegressor(loss='ls', learning_rate=0.1, n_estimators=3, max_depth=2).fit(x_train,y_train)
```

```python
pred_list = [1]
new_date = []
days=70
start = "5/21/20"
start_date = datetime.datetime.strptime(start, '%m/%d/%y')
for i in range(days):
    new_date.append((start_date+datetime.timedelta(days=i)).strftime('%m/%d/%y'))
    pred_list.append(model.predict(np.array(df_india["India"].max()+pred_list[-1]).reshape(-1,1))[0])
```

```python
plt.plot(new_date,pred_list)
```
